```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(100)
x_n = np.sin(0.2 * np.pi * n)

y_n = np.zeros(100)

for i in range(1, 100):
    y_n[i] = 0.7 * y_n[i-1] + x_n[i]

plt.figure(figsize=(10, 6))
plt.subplot(1, 1, 1)
plt.plot(n, x_n, label='Input Signal x[n]', color='blue')
plt.plot(n, y_n, label='System Output y[n]', color='pink')
plt.title('Linear Time-Invariant System Response')
plt.grid()
plt.legend()
plt.show()
```

```python
n = np.arange(100)
x_n = np.sin(0.2 * np.pi * n)


y_n = x_n ** 2


plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.scatter(n, x_n, color='blue', label='Input Signal x[n] = sin(0.2πn)', alpha=0.6)
plt.title('Scatter Plot')
plt.xlabel('Sample Number (n)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
n = np.arange(100)
x_n = np.sin(0.2 * np.pi * n)


y_n = x_n ** 2
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.bar(n, y_n, color='teal', label='Output Signal y[n] = (x[n])^2', alpha=0.6)
plt.title('bar plot')
plt.xlabel('Sample Number (n)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
n = np.arange(100)
x_n = np.sin(0.2 * np.pi * n)


y_n = x_n ** 2
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(n, x_n, label='Input Signal x[n]', color='black', linewidth=2)
plt.plot(n, y_n, label='Output Signal y[n]', color='blue', linewidth=2)
plt.title('Normal plot')
plt.xlabel('Sample Number (n)')
plt.ylabel('Amplitude')
plt.grid()
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
n_samples = 100
n = np.linspace(0, n_samples - 1, n_samples)
x_clean = 3 * np.sin(0.3 * np.pi * n)
noise = np.random.normal(0, 0.5, n_samples)
x_noisy = x_clean + noise

y = np.zeros(n_samples)
for i in range(1, n_samples):
    y[i] = 0.5 * y[i-1] + 0.5 * x_noisy[i]

plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(n, x_clean, label='Clean Input', color='blue')
plt.title('Clean Input')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(n, x_noisy, label='Noisy Input Signal', color='teal')
plt.title('Noisy Input Signal')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(n, y, label='System Output', color='black')
plt.title('System Output')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
