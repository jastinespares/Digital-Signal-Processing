```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000)
tau = 2

# Define original and shifted signals directly
y_t = 2 * np.sin(t)
y_t_shift = 2 * np.sin(t - tau)

# Plotting
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y_t, label="y(t) = 2 * sin(t)", color="blue")
plt.plot(t, y_t_shift, label="y(t - ) = 2 * sin(t - )", color="teal")
plt.title("Original and Time-Shifted Signal (Multiply by a Constant)")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend()
plt.show()
```

```python
t = np.linspace(0, 10, 1000) 
x_t = np.sin(t)
y_t = t * x_t
tau = 2


t_shifted = t - tau
x_t_shift = np.sin(t_shifted)
y_t_shift = (t_shifted)*x_t_shift

# Plotting
plt.figure(figsize=(10, 6))
plt.subplot(2,1,1)
plt.plot(t, y_t, label="y(t)= 2*x(t)", color="blue")
plt.xlabel('X')
plt.grid()
plt.plot(t, y_t_shift, label="y(t - ) = (t - ) * x(t - ) ", color="teal")
plt.title("Original and Time-Shifted Signal (Multiply by Time)")
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
t = np.linspace(0, 10, 1000) 
x_t = np.sin(t)
y_t = np.gradient(x_t, t)
tau = 2

t_shifted = t - tau
x_t_shift = np.sin(t_shifted)
y_t_shift = np.gradient(x_t_shift, t)

# Plotting
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, y_t, label="y(t) = d/dt*x(t)", color='blue')
plt.xlabel("X")
plt.plot(t, y_t_shift, label="y(t - ) = d/dt*x(t - )", color='teal')
plt.title("Original and Time-Shifted Signal (Differentiation)")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```
