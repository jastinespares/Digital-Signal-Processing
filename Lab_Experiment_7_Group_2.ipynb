```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(100)
clean_signal = np.sin(2 * np.pi * 0.05 * n)
noisy_signal = clean_signal + np.random.normal(0, 0.5, 100)
filtered_signal[i] = (noisy_signal[i] + noisy_signal[i-1] + noisy_signal[i-2]) / 3


plt.figure(figsize=(10, 6))
plt.plot(n,noisy_signal, label='Noisy input signal', color='blue')
plt.plot(n,filtered_signal, label='Filtered signal', color='red', linestyle='--')
plt.xlabel('Sample index')
plt.ylabel('Signal amplitude')
plt.title('Problem 1 :Digital Filtering in Signal Processing')
plt.legend(loc='upper right')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
n = np.arange(100)
base_price = 100  
stock_prices = base_price + np.random.normal(0, 2, len(n))
smoothed_prices = np.zeros_like(stock_prices)

for i in range(1, len(n)-1): 
    smoothed_prices[i] = (stock_prices[i-1] + stock_prices[i] + stock_prices[i+1]) / 3
smoothed_prices[0] = stock_prices[0]
smoothed_prices[-1] = stock_prices[-1]

plt.figure(figsize=(10, 6))
plt.title('Problem 2: Stock Price Prediction with Non-Causal Moving Average Filter')
plt.plot(stock_prices, label='Original Stock Prices', color='teal')
plt.plot(smoothed_prices, label='Smoothed Stock Prices', color='red',linestyle='--')
plt.xlabel('Day')
plt.ylabel('Stock Price')
plt.legend(loc='upper right')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
n = np.arange(100)
x = np.sin(0.1 * np.pi *  n)
d = 5

alpha = 0.6  
echo = np.zeros(100)
echo[d:] = alpha * x[:-d]  
x_with_echo = x + echo
y = np.copy(x_with_echo) 
y[d:] -= alpha * x[:-d]

plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(n, x_with_echo, label='Signal with Echo', color='teal')
plt.title("Problem: 3 Echo Cancellation in Telecommunications(Casual System)")
plt.ylabel('Amplitude')
plt.legend(loc='upper right')
plt.plot(n, y, label='Echo-cancelled Signal', color='red', linestyle='--')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend(loc='upper right')
plt.grid()
plt.tight_layout()
plt.show()
```
