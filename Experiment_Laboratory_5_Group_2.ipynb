```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000)
tau = 2

y_t = 2 * np.sin(t)
y_t_shift = 2 * np.sin(t - tau)

# Plotting
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, y_t, label="y(t) = 2 * sin(t)", color="teal")
plt.plot(t, y_t_shift, label="y(t - τ) = 2 * sin(t - τ)", color="blue")
plt.title("Original and Time-Shifted Signal (Multiply by a Constant)")
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.legend()
plt.show()
```

```python
t = np.linspace(0, 10, 1000) 
tau = 2

x_t = np.sin(t)
y_t = t * x_t


t_shifted = t - tau
x_t_shift = np.sin(t_shifted)
y_t_shift = (t_shifted)*x_t_shift

# Plotting
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, y_t, label="y(t)= 2*x(t)", color="blue")
plt.plot(t, y_t_shift, label="y(t - τ) = (t - τ) * x(t - τ) ", color="teal")
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Original and Time-Shifted Signal (Multiply by Time)")
plt.grid()
plt.legend()
plt.show()
```

```python
t = np.linspace(0, 10, 1000) 
x_t = np.sin(t)
y_t = np.gradient(x_t, t)
tau = 2

t_shifted = t - tau
x_t_shift = np.sin(t_shifted)
y_t_shift = np.gradient(x_t_shift, t)

# Plotting
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(t, y_t, label="y(t) = d/dt*x(t)", color='blue')
plt.plot(t, y_t_shift, label="y(t - τ) = d/dt*x(t - τ)", color='teal')
plt.xlabel("X")
plt.title("Original and Time-Shifted Signal (Differentiation)")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```
