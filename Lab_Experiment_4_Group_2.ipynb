```python
import numpy as np
import matplotlib.pyplot as pl
```

```python
N = np.linspace(-20, 20, 31)

def impulse(N):
   return np.where(N == 0, 1, 0)


plt.figure(figsize=(10, 8))
plt.plot(N, impulse(N), color='teal', label='Impulse Plot')
plt.stem(N, impulse(N), linefmt='coral', label='Impulse Stem')


plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
N = np.linspace(-10, 10, 100)
def ramp(X,Y,Z):
    return X*Y + Z

plt.figure(figsize=(10, 10))
plt.title('Ramp Signal')
plt.plot(N, ramp (N, 2, 0), label='m=1, b=0', color='teal')
plt.plot(N, ramp (N, 0.5, 0), label='m=0.5, b=0', color='darkRed')
plt.plot(N, ramp (N, -1, 0), label='m=1, b=2', color='coral')
plt.plot(N, ramp (N, -1.5, 3), label='m=1, b=2', color='Green' )

plt.axhline(color='k')
plt.axvline(color='k')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
N = np.linspace(-3, 3, 41)


plt.figure(figsize=(10, 8))
plt.title('Expontential Signal')
plt.stem(N,np.exp(N), linefmt='darkorange', label=r'$e^ns')
plt.stem(N,np.exp(-N), linefmt='teal', label=r'$e^(-n)$')
plt.stem(N,-np.exp(N), linefmt='coral', label=r'$-e^{n}$')
plt.stem(N,-np.exp(-N), linefmt='dodgerblue', label=r'$-e^(-n)$')



plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
N = np.linspace(-10, 10, 21)

def step(N):
    return np.where(N >=0, 1, 0)


plt.figure(figsize=(10, 4))
plt.title('Unit Signal')
plt.plot(N, step(N), color='teal', label='Step Plot')
plt.stem(N, step(N,), label='Step Stem')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
T = np.linspace(-2*np.pi, 2*np.pi, 100)
T2 = np.linspace(-2*np.pi, 4*np.pi, 200)

plt.figure(figsize=(10, 10))
plt.title('Amplitude Shift')

plt.plot(T, np.sin(T), label=r'$\sin \theta$', color='Red')
plt.plot(T2, 2*np.sin(T2), label=r'$2\sin \theta$')
plt.stem(T, np.sin(T), label=r'$\sin \theta$', linefmt='Blue')
plt.stem(T2, 2*np.sin(T2), label=r'$2\sin \theta$', linefmt='Gray')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Frequency Shift')

plt.plot(T, np.sin(T), label=r'$\sin \theta$', color="darkviolet")
plt.stem(T, np.sin(T), label=r'$\sin \theta$', linefmt='teal')
plt.plot(T, np.sin(2 * T), label=r'$\sin 2\theta$', color='darkorange')
plt.stem(T2, np.sin(2 * T2), label=r'$\sin 2\theta$', linefmt='Coral')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title('Phase Shift')

plt.plot(T, np.sin(T), label=r'$\sin \theta$', color='darkviolet')
plt.stem(T, np.sin(T), label=r'$\sin \theta$', linefmt='teal')
plt.plot(T2, np.sin(T2+np.pi/4), label=r'$\sin \theta + \frac{\pi}{2}$')
plt.stem(T2, np.sin(T2+np.pi/4), label=r'$\sin \theta + \frac{\pi}{2}$', linefmt='darkorange')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```

```python
def euler(theta):
    return np.exp(1j*theta)
```

```python
T = np.linspace(-4*np.pi, 4*np.pi, 400)

plt.figure(figsize=(10, 8))
plt.plot(T, euler(T).real, label='Real')
plt.plot(T, euler(T).imag, label='Imaginary', color='black')

plt.stem(T, euler(T).real, label='Real', markerfmt='go', basefmt='r')
plt.stem(T, euler(T).imag, label='Imaginary', markerfmt='bo', basefmt='g')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid()
plt.show()
```
